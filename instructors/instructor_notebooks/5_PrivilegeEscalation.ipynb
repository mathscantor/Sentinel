```python
# TODO
# 1. Using pd.read_csv(..., dtype=str), read in "../datasets/windows/csv/Security.csv" to a dataframe variable named 'df_security'
# 2. Filter events between the first time powershell was invoked and whoami /all was ran.
# 3. Filter EventData.ProcessName to only have powershell.exe

df_security = pd.read_csv("../datasets/windows/csv/Security.csv", dtype=str)
df_security_filtered = df_security[df_security["System.TimeCreated_attributes.SystemTime"] >= "2023-02-10T07:59:47.905061Z"]
df_security_filtered = df_security_filtered[df_security_filtered["System.TimeCreated_attributes.SystemTime"] <= "2023-02-10T08:06:38.231547Z"]
df_security_filtered = df_security_filtered[df_security_filtered["EventData.ProcessName"] == r"C:\Windows\System32\WindowsPowerShell\v1.0\powershell.exe"]
```
